```python
def stock_data(stock, factorList, startDate, endDate):
    data = w.wsd(stock, factorList, startDate, endDate,usedf = True)
    return data
```

```python
stockList = ["002415.SZ"]
startDate = "20200101"
endDate = "20201231"
factorList = ["open","close","volume"]
```

```python
data_all = pd.DataFrame()
for stock in stockList:
    data_temp = w.wsd(stock, factorList, startDate, endDate,usedf = True)[1].dropna()
    data_temp['Ticker'] = stock
    data_temp.index.name = 'Date'
    data_all = pd.concat([data_all, data_temp])
data_all = (data_all
            .reset_index()
            .set_index(['Date','Ticker'])
            .swaplevel())
```

```python
def find_large_pos_line(stock,data_all): 
    window_size = 22
    required_return = 0.04
    stock_data = data_all.loc[stock,['OPEN','CLOSE']]
    peak_timedf = pd.DataFrame()
    for date_idx in range(window_size,len(stock_data)):
        close_= stock_data.iloc[date_idx-window_size:date_idx]['CLOSE'].to_frame()
        if close_[close_ == close_.max()].dropna().index[0] == close_.iloc[-1].name: ## 判断 T 天收盘价是 (T-N,T) 天新高  改进措施用90th quantile
            
            open_ = stock_data.iloc[date_idx-window_size:date_idx]['OPEN'].to_frame()
            print(close_)
            print(close_.iloc[-1][0])
            print(open_.iloc[-1][0])
            if (((close_.iloc[-1][0]/open_.iloc[-1][0])-1) >= required_return): ## 判断 T 天收益率是否 大于 required_return
                peak_timedf = pd.concat([peak_timedf, stock_data.iloc[date_idx-1].to_frame().T])
    return peak_timedf.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
def check_next_day(stock,data_all,peak_timedf):
    required_return = .01
    ratio = .75
    long_timedf = pd.DataFrame()
    stock_data = data_all.loc[stock]
    for date in peak_timedf.index:
        
        if abs(stock_data.shift(-1).loc[date]['CLOSE']/stock_data.shift(-1).loc[date]['OPEN']-1) <= required_return:
            if stock_data.loc[date]['VOLUME'] * ratio > stock_data.shift(-1).loc[date]['VOLUME']:
                df_temp = stock_data.shift()['VOLUME'].to_frame()
                date_next = df_temp.where(df_temp==stock_data.loc[date]['VOLUME']).dropna().index[0]
                long_timedf = pd.concat([long_timedf,stock_data.loc[date_next].to_frame().T])
    return long_timedf
```

```python
def signal_long(stock,data_all):
    large_pos_df = find_large_pos_line(stock,data_all)
    long_timedf = check_next_day(stock,data_all,large_pos_df)
    return long_timedf
```
